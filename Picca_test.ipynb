“delta fields” — flux-transmission fluctuations δF = (F / ⟨F⟩) - 1.

```python
# ============================
# Step 0: Imports & Setup
# ============================
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
import fitsio
from astropy.table import Table
import picca.wedgize

# ============================
# Step 1: Define Correlation Class
# ============================
class Correlation:
    def __init__(self, data_file=None, fit_file=None):
        self.data = data_file
        if data_file:
            self.read_data(data_file)
        self.fit = fit_file
        if fit_file:
            self.read_fit(fit_file)

    def read_data(self, data_file):
        h = fitsio.FITS(data_file)
        self.da = h[1]['DA'][:]
        self.nb = h[1]['NB'][:]
        self.co = h[1]['CO'][:]
        self.rp = h[1]['RP'][:]
        self.rt = h[1]['RT'][:]
        hh = h[1].read_header()
        self.rpmin = hh['RPMIN']
        self.rpmax = hh['RPMAX']
        self.rtmin = 0
        self.rtmax = hh['RTMAX']
        self.nrp = hh['NP']
        self.nrt = hh['NT']
        h.close()
        self.r = np.sqrt(self.rp**2 + self.rt**2)
        self.read_dmat()
        self.read_metal_dmat()

    def read_dmat(self):
        if self.data:
            basedir, fname = os.path.split(self.data)
            dmat_file = fname.replace('cf','dmat').replace('-exp','')
            dmat_file = os.path.join(basedir, dmat_file)
            try:
                h = fitsio.FITS(dmat_file)
                self.dmat_file = dmat_file
                self.dmat = h[1]['DM'][:]
                h.close()
            except:
                print(f"MISSING DMAT FILE: {dmat_file}")

    def read_metal_dmat(self):
        if self.data:
            basedir, dmat_name = os.path.split(self.dmat_file)
            metal_file = 'metal_'+dmat_name.replace('.gz','')
            metal_file = os.path.join(basedir, metal_file)
            try:
                h = fitsio.FITS(metal_file)
                self.metal_dmat_file = metal_file
                self.metal_dmat = Table(h[2][:])
                h.close()
            except:
                print(f"MISSING METAL DMAT FILE: {metal_file}")

# ============================
# Step 2: Define plotting functions
# ============================
def plot_wedges(dataset, mu_values=[1.,0.95,0.8,0.5,0], power=2, rmin=10, rmax=180, label=None, limits=None, title=None):
    mu_low = mu_values[:-1]
    mu_high = mu_values[1:]
    plt.figure(figsize=(16,8))
    for i, (mumax, mumin) in enumerate(zip(mu_low, mu_high)):
        plt.subplot(2,2,i+1)
        for j,data in enumerate(dataset):
            b = picca.wedgize.wedge(mumin=mumin, mumax=mumax,
                                     rpmin=data.rpmin, rpmax=data.rpmax,
                                     rtmin=data.rtmin, rtmax=data.rtmax,
                                     nrt=data.nrt, nrp=data.nrp, absoluteMu=False)
            r, wedge_data, wedge_cov = b.wedge(data.da, data.co)
            y = wedge_data * r**power
            yerr = np.sqrt(np.diag(wedge_cov))*r**power
            p=plt.errorbar(r, y, yerr, fmt='o', label=label[j] if label else None)
            c=p[0].get_color()
            plt.xlabel(r"$r \, [h^{-1}\rm{Mpc}]$")
            plt.ylabel(r"$r^{}\xi(r)$".format(power))
            plt.title(r"${}<\mu<{}$".format(mumin,mumax), fontsize=16)
            plt.grid(linestyle='--')
            if data.fit is not None:
                r, wedge_model, _ = b.wedge(data.fit, data.co)
                ymod = wedge_model * r**power
                w = (r>=rmin) & (r<=rmax)
                plt.plot(r[w], ymod[w], linestyle='--', linewidth=2, color=c)
        if i==0 and label:
            plt.legend()
        if limits is not None:
            plt.ylim(limits[2*i], limits[2*i+1])
    if title is not None:
        plt.suptitle(title, fontsize=20)
    plt.tight_layout()
    plt.show()

def plot_corr_mat(dataset, power=2, label=None, vmin=None, vmax=None, title=None):
    nfigs = len(dataset)
    plt.figure(figsize=(6*nfigs,6))
    for i, data in enumerate(dataset):
        extent = [data.rtmin, data.rtmax, data.rpmin, data.rpmax]
        nrp,nrt = data.nrp, data.nrt
        r = data.r.reshape(nrp,nrt)
        mat = data.da.reshape(nrp,nrt)*r**power
        plt.subplot(1,nfigs,i+1)
        plt.imshow(mat, origin='lower', extent=extent, interpolation='nearest', cmap='seismic', vmin=vmin, vmax=vmax)
        cbar = plt.colorbar()
        cbar.set_label(r'$r\xi(r_{\parallel},r_{\perp})$')
        plt.xlabel(r'$r_\perp$ [Mpc/h]')
        plt.ylabel(r'$r_\parallel$ [Mpc/h]')
        if label: plt.title(label[i])
        plt.grid(True)
    if title: plt.suptitle(title, fontsize=16)
    plt.tight_layout()
    plt.show()

# ============================
# Step 3: Run PICCA CF on all Delta files
# ============================
delta_dir = "/global/cfs/cdirs/desicollab/users/lauracdp/photo-z_box/lya_mocks/mock_analysis/qq_desi_y5/skewers_desi_footprint.5/analysis-0/jura-0/raw/deltas_lya/Delta"
output_dir = "/global/u1/o/otilia/test"

delta_files = sorted(glob.glob(os.path.join(delta_dir,"*.fits*")))

for f in delta_files:
    base = os.path.basename(f).replace(".fits.gz","")
    output_file = os.path.join(output_dir, f"cf_{base}.fits.gz")
    
    print(f"Computing CF for {f}")
    
    # Run PICCA correlation function
    os.system(f"picca_cf.py --in-dir {f} --out {output_file} --r-min 0 --r-max 200 --dr 4 --z-min 1.9 --z-max 3.5 --nproc 8")

# ============================
# Step 4: Load & Plot
# ============================
cf_results = []
labels = []

for f in sorted(glob.glob(os.path.join(output_dir,"cf_*.fits.gz"))):
    cf_data = Correlation(data_file=f)
    cf_results.append(cf_data)
    labels.append(os.path.basename(f))

# Plot wedges
plot_wedges(cf_results, label=labels, title="3D CF Wedges")

# Plot correlation matrices
plot_corr_mat(cf_results, label=labels, power=1)
```
